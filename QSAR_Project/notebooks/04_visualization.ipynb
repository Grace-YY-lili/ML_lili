```python
# ==========================================
# File: 04_visualization.ipynb (EST Academic Style)
# ==========================================

import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

# 1. ä¿®æ­£è·¯å¾„
if 'notebooks' in os.getcwd(): os.chdir('..')

# æ¥æ”¶æ¨¡å‹åå­— (ç”¨äºä¿å­˜æ–‡ä»¶åï¼Œä¸ç”¨äºå›¾è¡¨æ ‡é¢˜ï¼Œä¿æŒä¸­æ€§)
try:
    current_name = model_tag
except NameError:
    current_name = "Model"

print(f"ğŸ¤– æ­£åœ¨ç»˜åˆ¶ {current_name} çš„å­¦æœ¯é£æ ¼å›¾è¡¨...")

# 2. è¯»å–æ•°æ®
try:
    df_train = pd.read_csv('results/predictions/train_predictions.csv')
    df_val = pd.read_csv('results/predictions/val_predictions.csv')
    df_test = pd.read_csv('results/predictions/final_predictions.csv')
except FileNotFoundError:
    raise FileNotFoundError("âŒ æ‰¾ä¸åˆ°ç»“æœæ–‡ä»¶ï¼è¯·å…ˆè¿è¡Œ 03ï¼")

# æå–æ•°æ®
y_train_true = df_train['Target_Log1o2']; y_train_pred = df_train['Predicted']
y_val_true = df_val['Target_Log1o2'];     y_val_pred = df_val['Predicted']
y_test_true = df_test['Target_Log1o2'];   y_test_pred = df_test['Predicted']

# è®¡ç®—æŒ‡æ ‡
r2_tr = r2_score(y_train_true, y_train_pred)
r2_val = r2_score(y_val_true, y_val_pred)
r2_te = r2_score(y_test_true, y_test_pred)
rmse_te = np.sqrt(mean_squared_error(y_test_true, y_test_pred))

# ========================================================
# 3. ç»˜å›¾æ ¸å¿ƒè®¾ç½® (å­¦æœ¯åŒ–æ”¹é€ )
# ========================================================

# A. å­—ä½“è®¾ç½® (ä¼˜å…ˆä½¿ç”¨ Arial, ç§‘ç ”é€šç”¨)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.it'] = 'Arial:italic'
plt.rcParams['mathtext.bf'] = 'Arial:bold'

# B. åˆ›å»ºç”»å¹… (æ­£æ–¹å½¢ï¼Œé«˜ DPI)
fig, ax = plt.subplots(figsize=(7, 7), dpi=300)

# C. åæ ‡è½´èŒƒå›´ (è‡ªåŠ¨è®¡ç®—å¹¶ç¨å¾®ç•™ç™½)
all_vals = np.concatenate([y_train_true, y_val_true, y_test_true])
min_val = np.min(all_vals) - 0.5
max_val = np.max(all_vals) + 0.5
ax.set_xlim(min_val, max_val)
ax.set_ylim(min_val, max_val)

# D. ç»˜åˆ¶å¯¹è§’çº¿ (åŸºå‡†çº¿)
# ä½¿ç”¨æ·±ç°è‰²è™šçº¿ï¼Œç¨å¾®ç»†ä¸€ç‚¹
ax.plot([min_val, max_val], [min_val, max_val], '--', color='#555555', linewidth=1.5, zorder=1)

# E. ç»˜åˆ¶æ•£ç‚¹ (å­¦æœ¯é…è‰²)
# Train: Steel Blue (é’¢è“), ä½é€æ˜åº¦, æ— è¾¹æ¡† -> ä½œä¸ºèƒŒæ™¯
ax.scatter(y_train_true, y_train_pred,
           c='#4682B4',        # Steel Blue
           alpha=0.25,         # éå¸¸æ·¡ï¼Œä¸æŠ¢æˆ
           s=50,
           edgecolors='none',
           marker='o',
           label=f'Train ($R^2={r2_tr:.2f}$)',
           zorder=2)

# Val: Sage Green (é¼ å°¾è‰ç»¿), ä¸­ç­‰é€æ˜åº¦ -> ä½œä¸ºä¸­é—´å±‚
ax.scatter(y_val_true, y_val_pred,
           c='#55A868',        # Muted Green
           alpha=0.6,
           s=60,
           edgecolors='none',
           marker='^',         # æ¢ä¸ªå½¢çŠ¶ (ä¸‰è§’å½¢) ä»¥ç¤ºåŒºåˆ«
           label=f'Val ($R^2={r2_val:.2f}$)',
           zorder=3)

# Test: Brick Red (ç –çº¢), é«˜é€æ˜åº¦, å¸¦ç»†è¾¹æ¡† -> è§†è§‰ç„¦ç‚¹
ax.scatter(y_test_true, y_test_pred,
           c='#C44E52',        # Brick Red
           alpha=0.85,
           s=80,
           edgecolors='#800000', # æ·±çº¢è‰²è¾¹æ¡†
           linewidth=0.8,
           marker='o',
           label=f'Test ($R^2={r2_te:.2f}, RMSE={rmse_te:.2f}$)',
           zorder=4)

# F. è£…é¥°ä¸æ ‡æ³¨ (å­¦æœ¯è§„èŒƒ)
# åæ ‡è½´æ ‡ç­¾ (åŠ ç²—ï¼Œå­—å·é€‚ä¸­)
ax.set_xlabel('Experimental $\log(1/O_2)$', fontsize=14, fontweight='bold', labelpad=10)
ax.set_ylabel('Predicted $\log(1/O_2)$', fontsize=14, fontweight='bold', labelpad=10)

# æ ‡é¢˜ (ä¸­æ€§ï¼Œå»é™¤æ¨¡å‹åå­—ï¼Œåªå†™ç‰©ç†å«ä¹‰)
ax.set_title('Experimental vs. Predicted Values', fontsize=16, pad=15)

# ç½‘æ ¼ (ä½ çš„è¦æ±‚ï¼šæå¼±)
ax.grid(True, linestyle=':', alpha=0.2, color='gray')

# åˆ»åº¦æœå†… (Inward ticks) - å¾ˆå¤šæœŸåˆŠå–œæ¬¢è¿™ä¸ªé£æ ¼
ax.tick_params(direction='in', length=5, width=1, colors='black', grid_alpha=0.2)

# è¾¹æ¡†åŠ ç²— (Box)
for spine in ax.spines.values():
    spine.set_linewidth(1.2)
    spine.set_color('black')

# å›¾ä¾‹ (æ”¾åœ¨å·¦ä¸Šè§’ï¼Œå»æ¡†æˆ–è€…ç®€æ˜“æ¡†)
ax.legend(loc='upper left', fontsize=11, frameon=True,
          fancybox=False, edgecolor='black', framealpha=0.8)

# ä¿æŒæ­£æ–¹å½¢æ¯”ä¾‹
ax.set_aspect('equal')

# ========================================================
# 4. ä¿å­˜
# ========================================================
os.makedirs('figures', exist_ok=True)

# è‡ªåŠ¨ç¼–å·
counter = 1
while True:
    fname = f"Academic_Plot_{current_name}_{counter}.png"
    full_path = os.path.join('figures', fname)
    if not os.path.exists(full_path): break
    counter += 1

plt.tight_layout()
plt.savefig(full_path, dpi=600, bbox_inches='tight') # 600 DPI å°åˆ·çº§æ¸…æ™°åº¦
print(f"ğŸ–¼ï¸å›¾å·²ç”Ÿæˆ: {full_path}")
plt.show()
```
